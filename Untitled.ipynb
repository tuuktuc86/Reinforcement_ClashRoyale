```python
import cv2
import pytesseract

def extract_numbers_from_image(image_path):
    # 이미지 읽기
    image = cv2.imread(image_path)
    
    # 이미지 전처리를 위한 그레이스케일 변환
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    
    img=Image.fromarray(gray_image)
    
    img1 = img.crop((90,110,140,140))
    
    img1 = np.array(img1)
    
    plt.imshow(img1)
    
    # Adaptive Thresholding을 사용하여 이미지 이진화
    threshold_image = cv2.adaptiveThreshold(img1, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY_INV, 11, 7)
    
    # 이미지를 흐리게 만들어 노이즈 제거
    blurred_image = cv2.medianBlur(threshold_image, 3)
    
    # 이미지에서 컨투어(윤곽선) 검출
    contours, hierarchy = cv2.findContours(blurred_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
    extracted_numbers = []
    
    for contour in contours:
        # 컨투어 영역의 크기 확인
        x, y, w, h = cv2.boundingRect(contour)
        
        # 컨투어 영역이 너무 작으면 무시
        if w < 10 or h < 10:
            continue
        
        # 컨투어 영역 추출
        digit_image = blurred_image[y:y+h, x:x+w]
        
        # 추출한 숫자 이미지를 흰 배경에 검은 숫자로 변환
        inverted_digit_image = cv2.bitwise_not(digit_image)
        
        # 추출한 숫자 이미지에서 텍스트 추출
        extracted_text = pytesseract.image_to_string(inverted_digit_image, config='--psm 10 --oem 3 -c tessedit_char_whitelist=0123456789')
        
        # 추출된 텍스트에서 숫자만 추출
        extracted_numbers.append(''.join(filter(str.isdigit, extracted_text)))
    
    return extracted_numbers

# 이미지 파일 경로 설정
image_path = '/home/jnu/Desktop/Reinforce/Royale/screenshot/cr_test/play_screen.png'

# 이미지에서 숫자 추출
numbers = extract_numbers_from_image(image_path)

# 추출된 숫자 출력
print(numbers)
```

```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

def preprocess_image(image):
    # 이미지 크기 조정 및 정규화
    image = cv2.resize(image, (28, 28))
    image = image.astype('float32') / 255.0
    
    # 모델 입력 형태로 변환 (배치 차원 추가)
    image = np.expand_dims(image, axis=0)
    
    return image

def extract_numbers_from_image(image_path):
    # 이미지 읽기
    image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    img=Image.fromarray(image)
    
    img1 = img.crop((90,110,140,140))
    
    img1 = np.array(img1)
    
    plt.imshow(img1)
    # 이미지 전처리
    preprocessed_image = preprocess_image(img1)
    
    # MNIST 모델 로드
    model = load_model('mnist_model.h5')
    
    # 숫자 인식 예측
    predictions = model.predict(preprocessed_image)
    
    # 예측 결과에서 가장 높은 확률 값을 가진 클래스 추출
    predicted_digits = np.argmax(predictions, axis=1)
    
    return predicted_digits

# 이미지 파일 경로 설정
image_path = '/home/jnu/Desktop/Reinforce/Royale/screenshot/cr_test/play_screen.png'

# 이미지에서 숫자 추출
numbers = extract_numbers_from_image(image_path)

# 추출된 숫자 출력
print(numbers)
```

```python
import gym
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.distributions import Categorical
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import torchvision.transforms as transforms

#Hyperparameters
learning_rate = 0.0005
gamma         = 0.98
lmbda         = 0.95
eps_clip      = 0.1
K_epoch       = 3
T_horizon     = 20
```

```python



class PPO(nn.Module):
    def __init__(self):
        super(PPO, self).__init__()
        self.data = []
        
        self.conv1 = nn.Sequential(nn.Conv2d(3, 16, 3, stride=2, padding=1),
                                   nn.BatchNorm2d(16),
                                   nn.ReLU(inplace=True))
        self.conv2 = nn.Sequential(nn.Conv2d(16, 64, 3, stride=2, padding=1),
                                   nn.BatchNorm2d(64),
                                   nn.ReLU(inplace=True))
        self.conv3 = nn.Sequential(nn.Conv2d(64, 256, 3, stride=2, padding=1),
                                   nn.BatchNorm2d(256),
                                   nn.ReLU(inplace=True))
        self.actor_linear1 = nn.Sequential(nn.Linear(1745920, 256),
                                          nn.ReLU(inplace=True),
                                          nn.Linear(256,1))
        self.actor_linear2 = nn.Sequential(nn.Linear(1745920, 256),
                                          nn.ReLU(inplace=True),
                                          nn.Linear(256,4))
        self.optimizer = optim.Adam(self.parameters(), lr=learning_rate)

    def pi(self, x, softmax_dim = 0):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.actor_linear2(x.reshape(x.size(0), -1))
        #prob = F.softmax(x, dim=softmax_dim)
        return x
    
    def v(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        v = self.actor_linear1(x.reshape(x.size(0), -1))
        return v
      
    def put_data(self, transition):
        self.data.append(transition)
        
    def make_batch(self):
        s_lst, a_lst, r_lst, s_prime_lst, prob_a_lst, done_lst = [], [], [], [], [], []
        for transition in self.data:
            s, a, r, s_prime, prob_a, done = transition
            
            s_lst.append(s)
            a_lst.append([a])
            r_lst.append([r])
            s_prime_lst.append(s_prime)
            prob_a_lst.append([prob_a])
            done_mask = 0 if done else 1
            done_lst.append([done_mask])
            
        s,a,r,s_prime,done_mask, prob_a = torch.tensor(s_lst, dtype=torch.float), torch.tensor(a_lst), \
                                          torch.tensor(r_lst), torch.tensor(s_prime_lst, dtype=torch.float), \
                                          torch.tensor(done_lst, dtype=torch.float), torch.tensor(prob_a_lst)
        self.data = []
        return s, a, r, s_prime, done_mask, prob_a
        
    def train_net(self):
        s, a, r, s_prime, done_mask, prob_a = self.make_batch()

        for i in range(K_epoch):
            td_target = r + gamma * self.v(s_prime) * done_mask
            delta = td_target - self.v(s)
            delta = delta.detach().numpy()

            advantage_lst = []
            advantage = 0.0
            for delta_t in delta[::-1]:
                advantage = gamma * lmbda * advantage + delta_t[0]
                advantage_lst.append([advantage])
            advantage_lst.reverse()
            advantage = torch.tensor(advantage_lst, dtype=torch.float)

            pi = self.pi(s, softmax_dim=1)
            pi_a = pi.gather(1,a)
            ratio = torch.exp(torch.log(pi_a) - torch.log(prob_a))  # a/b == exp(log(a)-log(b))

            surr1 = ratio * advantage
            surr2 = torch.clamp(ratio, 1-eps_clip, 1+eps_clip) * advantage
            loss = -torch.min(surr1, surr2) + F.smooth_l1_loss(self.v(s) , td_target.detach())

            self.optimizer.zero_grad()
            loss.mean().backward()
            self.optimizer.step()
```

```python

        
def main():
    env = gym.make('CartPole-v1')
    model = PPO()
    score = 0.0
    print_interval = 20

    for n_epi in range(10000):
        s, _ = env.reset()
        done = False
        while not done:
            for t in range(T_horizon):
                prob = model.pi(torch.from_numpy(s).float())
                m = Categorical(prob)
                a = m.sample().item()
                s_prime, r, done, truncated, info = env.step(a)

                model.put_data((s, a, r/100.0, s_prime, prob[a].item(), done))
                s = s_prime

                score += r
                if done:
                    break

            model.train_net()

        if n_epi%print_interval==0 and n_epi!=0:
            print("# of episode :{}, avg score : {:.1f}".format(n_epi, score/print_interval))
            score = 0.0

    env.close()

if __name__ == '__main__':
    main()
```
